# 1. Test Random Environment with OpenAI Gym

```python
from gym import Env
from gym.spaces import Discrete, Box
import numpy as np
import random
print(Box(low=np.array([0]), high=np.array([100])))
```

```python
# 기본 ENV 템플렛 : step, render, reset
class ShowerEnv(Env):
    def __init__(self):
        # action_space : 모델이 취할 수 있는 행동들의 집합 : Actions we can take = down, stay, up : 0, 1, 2
        self.action_space = Discrete(3) 

        # observation_space : 반환할 state의 형식과 범위의 정의  : 여기서는 Temperature array로 사용. 즉 self.state는 0이상 100이하의 값을 가질 수 있음.
        self.observation_space = Box(low=np.array([0]), high=np.array([100])) 
        
        # state : 환경의 상태 : Set start temp
        self.state = 38+ random.randint(-3, 3)

        # Set episode length
        self.episode_length = 60
        
    def step(self, action): # action(여기서는 0, 1, 2중 하나)을 취했을 때, state의 변화와 그에 따른 보상을 책정하는 부분
        self.state += action - 1 # Apply action : -1을 한 이유는 action=0이면 온도를 내려야하니 0 - 1 로 온도 1도를 내리는 것으로 이해
                                 # action = 1 : 1 - 1 = 0 으로 온도 유지
                                 # action = 2 : 2 - 1 = 1 으로 온도 상승
        self.episode_length -= 1 # Reduce episode length

        # Calculate reward
        if self.state >= 37 and self.state <= 39:
            reward = 1
        else:
            reward = -1

        # Check if episode is done
        if self.episode_length <= 0:
            done = True
        else:
            done = False

        # Apply temperature noise : 실제 환경과 유사하게 noise를 추가해서 모델 학습에 추가했음
        self.state += random.randint(-1, 1)

        # Set placeholder for info
        info = {}

        return self.state, reward, done, info
        
    def render(self): # visualization for the enviornment ex) pygame
        pass
    def reset(self): # 한 episodse를 시작할 때 attributes(actoin_space, observation_space, state, reward)를 초기화 하는 함수
        self.state = 38 + random.randint(-3, 3) # Reset shower temperature
        self.episode_length = 60 # Reset episode length
        return self.state
```

```python
env = ShowerEnv()
```

```python
env.action_space.sample() # 임의의 action을 선택
```

```python
env.observation_space.sample() #  Box의 범위 내에서 무작위로 샘플을 생성, RL 초기 상태를 설정하거나 테스트 목적으로 사용
```

```python
# Testing a Custom OpenAI Env

episodes = 10
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 

    # 1. while문에 들어가게 되면 ShowerEnv에서 설정한 episode_length만큼 반복문 실행됨
    # 2. episode_length만큼 실행이 끝나면 while문 벗어나 for문으로 돌아와 episodes만큼 1. 을 다시 수행
    # 즉 Episode:1에서 Score가 -2인 것은 self.state가 37이상 39이하 일때보다 아닐때가 2번 더 많았다는 뜻.
    while not done:
        #env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
```

# 2. Create a Deep Learning Model with Keras

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers.legacy import Adam
```

```python
states = env.observation_space.shape
actions = env.action_space.n
```

```python
print(f"states : {states} actions : {actions}")
```

```python
def build_model(states, actions):
    model = Sequential()    
    model.add(Dense(32, activation='relu', input_shape=states))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(actions, activation='linear'))
    return model
```

```python
del model # for 'Sequential' object has no attribute '_compile_time_distribution_strategy' error 
```

```python
model = build_model(states, actions)
```

```python
model.summary()
```

# 3. Build Agent with Keras-RL

```python
import tensorflow 
from keras import __version__
tensorflow.keras.__version__ = __version__

from rl.agents import DQNAgent
from rl.policy import BoltzmannQPolicy
from rl.memory import SequentialMemory
```

```python
def build_agent(model, actions):
    policy = BoltzmannQPolicy()
    memory = SequentialMemory(limit=50000, window_length=1)
    dqn = DQNAgent(model=model, memory=memory, policy=policy, 
                  nb_actions=actions, nb_steps_warmup=10, target_model_update=1e-2)
    return dqn
```

```python
# Model Training

dqn = build_agent(model, actions)
dqn.compile(Adam(lr=1e-3), metrics=['mae'])
dqn.fit(env, nb_steps=500000, visualize=False, verbose=1) # np_steps을 변경하여 모델의 성능을 향상시킬 수 있음.
```

```python
scores = dqn.test(env, nb_episodes=100, visualize=False)
print(np.mean(scores.history['episode_reward']))
```
